```python
import pandas as pd
from sklearn.model_selection import train_test_split
import csv
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
import seaborn as sn
```

```python
# Load data frame with pandas (pd) uncomment the one you want to see
# training_data = pd.read_csv("./diabetes/training_data_diabetes.csv")
training_data = pd.read_csv("./diabetes/training_data(no_pre-diabetes).csv")
```

Determine for each class the occurence percentage and thus the potential accuracy of a major class classification

```python
# Get occurence percentage for each label represented in the dataset. Going from big to small.
label_percentage_total = training_data["Diabetes_012"].value_counts(ascending=False, dropna=True, normalize=True)
print(label_percentage_total)

# Get most represented class percentage
accuracy_major_class_classification = label_percentage_total[0]
most_represented_label = training_data["Diabetes_012"].value_counts().idxmax()

#Print outcome
print("Most represented label: ", most_represented_label)
print("Expected accuracy of major class classification: ", accuracy_major_class_classification * 100, "%")
```

```python
for column in training_data:
    training_data.pivot(columns="Diabetes_012", values=column).plot.hist()
    plt.title(column)
    plt.show()
```

```python
correlation_matrix = training_data.corr()
```

```python
corr_heatmap = sn.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
sn.set(font_scale= 0.25)
plt.title("Correlation matrix Diabetes Health Data (no pre-diabetes)")
plt.show()
```
