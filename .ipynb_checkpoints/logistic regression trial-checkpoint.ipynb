```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data_pre = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")
# Split the dataset into train and test sets
X_pre = data_pre.drop('Diabetes_012', axis = 1)
y_pre = data_pre['Diabetes_012']
# The ratio of training set to test set is 90% and 10%
X_pre_train, X_pre_test, y_pre_train, y_pre_test = train_test_split(X_pre, y_pre, train_size = 0.9, random_state=0)
# Build a Logistic Classifier from the Training Set
logistic_pre = sm.MNLogit(y_pre_train,X_pre_train).fit()
logistic_pre.summary()
```

```python
data_balance = pd.read_csv("diabetes_binary_5050split_health_indicators_BRFSS2015.csv")
# Split the dataset into train and test sets
X_balance = data_balance.drop('Diabetes_binary', axis = 1)
y_balance = data_balance['Diabetes_binary']
# The ratio of training set to test set is 90% and 10%
X_balance_train, X_balance_test, y_balance_train, y_balance_test = train_test_split(X_balance, y_balance, train_size = 0.9, random_state=0)
# Build a Logistic Classifier from the Training Set
logistic_balance = sm.Logit(y_balance_train,X_balance_train).fit()
logistic_balance.summary()
```

```python
data = pd.read_csv("diabetes_binary_health_indicators_BRFSS2015.csv")
# Split the dataset into train and test sets
X = data.drop('Diabetes_binary', axis = 1)
y = data['Diabetes_binary']
# The ratio of training set to test set is 90% and 10%
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.9, random_state=0)
# Build a Logistic Classifier from the Training Set
logistic = sm.Logit(y_train,X_train).fit()
logistic.summary()
```

```python
print(logistic_pre.aic)
print(logistic_balance.aic)
print(logistic.aic)
```
