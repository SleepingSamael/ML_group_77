```python
import numpy as np
import pandas as pd
import sns as sns
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# create a dict to store the scores of each model
models_evalutions = {'Model':[],
                     'Accuracy':[],
                     'Precision':[],
                     'Recall':[],
                     'F1 score':[]}
```

```python
# Reading the data
data = pd.read_csv("../diabetes/training_data(no_pre-diabetes).csv")
```

```python
# Splitting features and target
x = data.iloc[:,1:]
y = data.Diabetes_012
# Splitting training dataset
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.1)
```

```python
# First KNN with hyperparameter 5
knn = KNeighborsClassifier(5)
knn.fit(x_train,y_train)

y_predicted = knn.predict(x_val)
```

```python
# Show accuracy
print('accuracy ', accuracy_score(y_val, y_predicted))
```

```python
# Check if picking up 8 important features improve the accuracy
essentials = ["HighBP","HighChol","BMI","HeartDiseaseorAttack","GenHlth","PhysHlth","DiffWalk","Age"]
knn.fit(x_train[essentials],y_train)
y_predicted_ess = knn.predict(x_val[essentials])
```

```python
# Accuracy didn't improve
print("accuracy",accuracy_score(y_val,y_predicted_ess))
```

```python
# Check if higher hyperparameter
knn_2 = KNeighborsClassifier(15)
knn_2.fit(x_train,y_train)

y_predicted_more = knn_2.predict(x_val)
```

```python
# No significant improvement
print('accuracy ', accuracy_score(y_val, y_predicted_more))
```

```python
"""
### Aoming added
WARNING:Will take some time, not necessary to run many times
"""
# visualize the plot between accuracy and K value
acc = []
from sklearn import metrics
for i in range(1,40):
    knn_trial = KNeighborsClassifier(n_neighbors = i)
    knn_trial.fit(x_train,y_train)
    yhat = knn_trial.predict(x_val)
    acc.append(metrics.accuracy_score(y_val, yhat))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),acc,color = 'blue',linestyle='dashed',
         marker='o',markerfacecolor='red', markersize=10)
plt.title('accuracy vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy')
print("Maximum accuracy:-",max(acc),"at K =",acc.index(max(acc)))
```

```python
"""
### Aoming added
WARNING:Will take some time, not necessary to run many times
"""
# visualize the plot between error rate and K value
error_rate = []
for i in range(1,40):
 knn = KNeighborsClassifier(n_neighbors=i)
 knn.fit(x_train,y_train)
 pred_i = knn.predict(x_val)
 error_rate.append(np.mean(pred_i != y_val))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed',
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
```

We can chose the best K value according to these two plots:
(Elbow method? or other methods)
Notice that larger K costs more

```python
# Try multiple hyperparameters to find the best K
for i in range(6,15):
    knn_trial = KNeighborsClassifier(i)
    knn_trial.fit(x_train,y_train)
    y_predicted_trial = knn_trial.predict(x_val)
    print("accuracy of hyperparamater {}".format(i), accuracy_score(y_val,y_predicted_trial))
```

```python
# Using KNN  train on training set, and Test on testing set
# Using K = 29 as example
knn = KNeighborsClassifier(n_neighbors=29)
knn.fit(x_train, y_train)
print("The score for kNN without cross val and without using smote is")
print("Training set: {:6.2f}%".format(100*knn.score(x_train, y_train)))
print("Validation set: {:6.2f}%".format(100*knn.score(x_val, y_val)))
# Reading the data
test_data = pd.read_csv("../diabetes/testing_data(no_pre-diabetes).csv")
x_test = data.iloc[:,1:]
y_test = data.Diabetes_012
print("Test set: {:6.2f}%".format(100*knn.score(x_val, y_val)))
```

```python
#test the model
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
#prediction
test_pred=knn.predict(x_test)
#Accuracy
confusion_hard = confusion_matrix( y_test, test_pred)
accuracy = accuracy_score(y_test , test_pred)
precision = precision_score(y_test , test_pred, pos_label = 2.0)
recall = recall_score(y_test , test_pred,pos_label = 2.0)
f1 = f1_score(y_test,test_pred, pos_label = 2.0)
print('\nKNN Accuracy for validation set=: {0:.4f}, \nprecision: {1:.4f}, \nrecall: {2:.4f},\
\nF1: {3:.4f}'.format(accuracy, precision, recall, f1))

models_evalutions['Model'].append("KNN after balance our target's labels")
models_evalutions['Accuracy'].append(accuracy)
models_evalutions['Precision'].append(precision)
models_evalutions['Recall'].append(recall)
models_evalutions['F1 score'].append(f1)
```
